```
import pandas as pd
import numpy as np
from sklearn.cluster import MeanShift
```

```
df = pd.read_csv('/content/drive/My Drive/datascience/checkins.csv')
```

```
coords = df[['latitude', 'longitude']]
```

```
coords = coords[:100000]
```

```
%%time
clustering = MeanShift(bandwidth=0.1, min_bin_freq = 15).fit(coords)
```

```
centers = clustering.cluster_centers_
```

```
centers = list(zip(clustering.cluster_centers_, clustering.labels_))
```

```
officces = [[33.751277, -118.188740], [33.751277, -118.188740], [25.867736, -80.324116], [51.503016, -0.075479], [52.378894, 4.885084],[39.366487, 117.036146],[-33.868457, 151.205134]]
```

```
from scipy.spatial.distance import euclidean
```

```
dists = []
i = 0
j = 0
for officce in officces:
  for center in centers:
    d = euclidean(officce, center[0])
    t = list((d, center[0]))
    dists.append(t)
    j+=1
  i+=1
```
